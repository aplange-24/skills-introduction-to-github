```python
# prompt: convert dict to dataframe

import pandas as pd
import io

# Assuming 'uploaded' is a dictionary where keys are filenames and values are file contents
# Example: uploaded = {'data.csv': b'col1,col2\n1,2\n3,4'}

for fn in uploaded.keys():
  if fn.endswith('.csv'):  # Assuming you want to convert CSV files
    df = pd.read_csv(io.BytesIO(uploaded[fn]))
    print(df)
    break  # Process only the first CSV file
  elif fn.endswith('.xlsx') or fn.endswith('.xls'):
    df = pd.read_excel(io.BytesIO(uploaded[fn]))
    print(df)
    break # Process only the first Excel file
  elif fn.endswith('.json'):
    df = pd.read_json(io.BytesIO(uploaded[fn]))
    print(df)
    break #Process only the first json file
  else:
    print(f"Unsupported file type: {fn}. Please provide a CSV, xlsx or json file.")
    # Handle other file types or raise an exception as needed.
```

```python
# import calendar to change month names to String Names
import calendar
#Extract year aand month for grouping
#Keep the original date time column as 'Months_dt'
df['Months_dt'] = pd.to_datetime(df['Months'])
df['Year'] = df['Months_dt'].dt.year
df['Month'] = df['Months_dt'].dt.month
df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
df.head()
```

```python
# d_months = df['Months']
# d_months = d_months.to_frame()
# d_months.head()

# #d_months.head()


dfs = df[['gasoline','gasoil','Month','Year']]
dfs.head()
```

```python
# Group gasoline losses by Year and Month
monthly_gasoline_comparison = df.groupby(['Year', 'Month'])['gasoline'].sum().unstack(level=0)
# arrange months in chronological order in months not alphabetical

# Display the grouped data for comparison
month_order = [calendar.month_abbr[i] for i in range(1, 13)]  # Create a list of month abbreviations in order
monthly_gasoline_comparison = monthly_gasoline_comparison.reindex(month_order)

monthly_gasoline_comparison
```

```python
# Group gasoil losses by Year and Month
monthly_gasoil_comparison = df.groupby(['Year', 'Month'])['gasoil'].sum().unstack(level=0)
# arrange months in chronological order in months not alphabetical

# Display the grouped data for comparison
# month_order = [calendar.month_abbr[i] for i in range(1, 13)]  # Create a list of month abbreviations in order
monthly_gasoil_comparison = monthly_gasoil_comparison.reindex(month_order)

monthly_gasoil_comparison
```

```python
import matplotlib.pyplot as plt
# Re-plot the month-by-month gasoline comparison with chronological order
plt.figure(figsize=(12, 8))
monthly_gasoline_comparison.plot(kind='bar', figsize=(14, 8), width=0.8, colormap='viridis')

# Add chart title and labels
plt.title('Month-by-Month Gasoline Loss Comparison (2021-2024)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Gasoline Losses', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Year', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
